```python
order_amount = input("Enter the order amount: ")
if not order_amount.isdigit():
    print("Please enter a valid number for the order amount.")
else:
    order_amount = int(order_amount)
    if order_amount >= 2000:
        amount = order_amount * 0.85
        print(f"The total amount after discount is: {amount}")
    elif order_amount >= 1500 and order_amount < 2000:
        amount = order_amount * 0.90
        print(f"The total amount after discount is: {amount}")
    elif order_amount >= 1000 and order_amount < 1500:  
        amount = order_amount * 0.95
        print(f"The total amount after discount is: {amount}")
    else:
        amount = order_amount
        print(f"No discount applicable. The total amount is: {amount}")
```
