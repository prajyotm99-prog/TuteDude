```python
orders = [1200,2500,800,1750,3000]
discount = []
price = []
```

```python
for i in orders:
    if not isinstance(i, (int, float)):
        discount.append(0)
        price.append(i)
    else:
        if i >= 2000:
            price.append(i * 0.85)
            discount.append(i * 0.15)
        elif i < 2000 and i >= 1500:
            price.append(i * 0.90)
            discount.append(i * 0.10)
        elif i >= 1000 and i < 1500:
            price.append(i * 0.95)
            discount.append(i * 0.05)
        else:
            price.append(i)
            discount.append(0)
            
        
```

```python
discount
```

```python
table = tuple(zip(orders, discount, price))
print(table)
    
```

```python
sum =  0
for p in table:
    if isinstance(p[2], (int, float)):
        sum = sum + p[2]
print(f"The total amount after discount is: {sum}")
```

```python
num = 0
for p in table:
    if isinstance(p[1], (int, float)):
        if p[1] > 0:
            num = num + 1
print(f"The total number of orders that received a discount is: {num}")
```
