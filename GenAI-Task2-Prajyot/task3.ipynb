```python
name, price = [], []
choice = '1'
```

```python
while choice != 'q':
    if choice == '1':
        name.append(input("Enter the name of the item: "))
        price.append(float(input("Enter the price of the item: ")))
    elif choice == '2':
        print("Items in the shopping cart:")
        for i in range(len(name)):
            print(f"{name[i]}: ${price[i]:.2f}")
    elif choice ==  '3':
        print(f"the total amount is {sum(price)}")
    elif choice == 'q':
        break
    else:
        continue
    choice = str(input("Do you want to add another item? (Enter 1 to add; 2 to view; 3 to Amount;'q' to quit): "))
```
