```python
product = ["apple", "potato","trimmer","tv","icecream","laptop"]
category = ["fruit", "vegetable", "tool", "electronics", "dessert", "electronics"]
price = [100.12, 50.87, 200, 500, 150.97, 800]
```

```python
dict_product_price = dict(zip(product, price))
print(dict_product_price)
```

```python
dict_product_price["tv"] = 450
print(dict_product_price)
```

```python
dict_product_price.pop("icecream")
print(dict_product_price)
```

```python
sum = 0
for key, value in dict_product_price.items():
    sum += value
print(f"Sum of all product prices: {sum}")
average_price = sum / len(dict_product_price)
print(f"Average price of products: {average_price}")
```

```python
product = max(dict_product_price.values()) * min(dict_product_price.values())
print(f"Product of maximum and minimum price: {product}")
   
```
