```python
product = ["apple", "potato","trimmer","tv","icecream","laptop"]
category = ["fruit", "vegetable", "tool", "electronics", "dessert", "electronics"]
price = [100, 50, 200, 500, 150, 800]
```

```python
catalog = tuple()
for p, c, pr in zip(product, category, price):
    new = ((p,c,pr),)
    catalog += new
print(catalog)
```

```python
category_to_products = {}
for p, c in zip(product, category):
    if c not in category_to_products:
        category_to_products[c] = []
    category_to_products[c].append(p)
print(category_to_products)
```

```python
counts = {}
for prod, cat, pr in catalog:
    if cat in counts:
        counts[cat] += 1
    else:
        counts[cat] = 1
max_category = None
max_count = 0

for cat, count in counts.items():
    if count > max_count:
        max_count = count
        max_category = cat
print(f"Products in the largest category ('{max_category}'):")
for prod, cat, pr in catalog:
    if cat == max_category:
        print(f"- {prod}")
```
