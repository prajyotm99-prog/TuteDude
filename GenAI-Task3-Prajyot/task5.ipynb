```python
prices = [100,250,400,1200,50,1500,2000,500]
```

```python
def gre(num):
    if num > 500:
        return True
    return False
def less(num):
    if num <= 500:
        return True
    return False
```

```python
new1 = list(filter(gre, prices))
new2 = list(filter(less, prices))
print(f"greater than 500 {new1}")
print(f"lesser than 500 {new2}")
```
