```python
prices_list = [100, 250, 400, 1200, 50, 1500, 2000, 500]

add_price = lambda prices, price: list(map(lambda x: x + price, prices))
ave = lambda prices: sum(prices) / len(prices)
high = lambda prices: max(prices)

while True:
    choice = input(
        "\n1. Add price\n2. Average\n3. Highest\nq. Quit\nEnter choice: "
    )

    if choice == '1':
        price = int(input("Enter the price to add: "))
        print(add_price(prices_list, price))

    elif choice == '2':
        print("Average:", ave(prices_list))

    elif choice == '3':
        print("Highest:", high(prices_list))

    elif choice == 'q':
        break

    else:
        print("Invalid choice")
```
