```python
prices = {"Mouse" : 500 , "Keyboard" : 800, "Monitor" : 7000, "Pendrive" : 400, "Camera" : 5000}
```

```python
for i in prices.items():
    print(i)
```

```python
dis = float(input("Enter discount percentage"))
total_discounted_price = 0
total_items = len(prices)
```

```python
with open("discount_report.txt", "w") as f:
    for product, original_price in prices.items():
        discounted_price = original_price * (1 - dis/ 100)
        total_discounted_price += discounted_price
        f.write(f"{product} | {original_price} | {discounted_price:.2f}\n")
    f.write(f"Average Discounted Price: {avg_discounted_price:.2f}\n")
    avg_discounted_price = total_discounted_price / total_items
    f.write(f"Total Items: {total_items}\n")
```
