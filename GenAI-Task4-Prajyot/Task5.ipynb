```python
names = ["Laptop", "Mouse", "Keyboard", "Monitor"]
prices = [1200, 25, 45, 300]
```

```python
with open('sales_price.txt', 'w') as sp:
    for name, price  in zip(names, prices):
        sp.write(f"{name} : {price} \n") 
    
```

```python
with open('sales_price.txt', 'r') as f:
    content = f.read()
print(content)
```
